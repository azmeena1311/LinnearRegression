```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Reading the data from the csv
data = pd.read_csv('Data_linear_regression.csv')
print(data)
```

```python
#Analyzing the data
data.plot(kind='scatter', x='area',y='prices')
data.plot(kind = 'box')
```

```python
#checking the corelation between the columns
print(data.corr())
```

```python
#preparing the data to feed the Linear model
Area=pd.DataFrame(data['area'])
Prices=pd.DataFrame(data['prices'])
```

```python
#Linear Regression model
lm=LinearRegression()
model=lm.fit(Area,Prices)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
#Evaluating the model
print(model.score(Area,Prices))
```

```python
#Predicting new values using the trained model
New_area=[[8000]]
New_pred =model.predict(New_area)
print(New_pred)
```

```python
#Predict multiple new values
Values=[[5000],[2900],[4300]]
X=pd.DataFrame(Values)
Y=model.predict(X)
Y=pd.DataFrame(Y)
df=pd.concat([X,Y],axis=1,keys=['Area_New','Predicted_Prices'])
print(df)
```

```python
#Visualizing the results
data.plot(kind='scatter',x='area',y='prices')
```

```python
#Plotting the regression line
plt.plot(Area,model.predict(Area),color='red',linewidth=4)
```

```python
#plotting the predicted value
plt.scatter(X,Y,color='black')
plt.show()
```
